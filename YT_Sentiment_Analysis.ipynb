##huggingface pipelining

```python
from transformers import pipeline
```

```python
pipe_Image2Text = pipeline(
    "image-to-text", model="Salesforce/blip-image-captioning-large", device = 0) # why device = 0?

pipe_GrammarCorrection = pipeline(
    "text2text-generation", model="pszemraj/flan-t5-large-grammar-synthesis", device = 0)

pipe_SentimentAnalysis = pipeline(
    "text-classification", model="lxyuan/distilbert-base-multilingual-cased-sentiments-student", return_all_scores = True, device = 0) # what is return_all_scores?
```

```python
descriptions = []

def imgToBase64(path : str) -> str:
    try:
      img = Image.open(path)
    except Exception as e:
      print(f"{path} is invalid , image doesn't exist")
      img = Image.open("/content/0.jpg")

    byte_arr = io.BytesIO()
    img.save(byte_arr, format='PNG')
    byte_arr = byte_arr.getvalue()
    return str(base64.b64encode(byte_arr).decode('utf-8'))

def imgPath(index: str) -> str:
  return f"/content/{index}.jpg"


def descriptionCleaner(output : list[dict]) -> str:
  return output[0]['generated_text']

for i in range(len(ids)):
  descriptions.append(descriptionCleaner(pipe_Image2Text(imgToBase64(imgPath(i)))))




descriptions
```

```python
def sentimentCleaner(output : list[list[dict]]) -> dict:

  sentiment = {
      "label" : "",
      "score" : float('-inf')
  }


  for dict in output[0]:
    if dict['score'] > sentiment['score']:
      sentiment['score'] = dict['score']
      sentiment['label'] = dict['label']


  return sentiment


def correctionCleaner(output : list[dict] ) -> str:
  return output[0]['generated_text']
```

```python
sentiments = []
for i in range(len(descriptions)):
  descriptions[i] = correctionCleaner(pipe_GrammarCorrection(descriptions[i]))
  sentiments.append(sentimentCleaner(pipe_SentimentAnalysis(descriptions[i])))
```

```python
def mapper(score, min , max):

    mapped_value = (score) * (max - min) + min

    return mapped_value
```

```python
from fastapi import FastAPI
import nest_asyncio
from pyngrok import ngrok
import uvicorn
from pydantic import BaseModel

app = FastAPI()

class QueryRequest(BaseModel):
  query: str

class SentimentResponse(BaseModel):
  sentiment: dict


@app.post('/analyze', response_model = SentimentResponse) # decorator specifying get, post and actual endpoint
async def Thumbnail_Sentiment(request:QueryRequest):

  driver = webdriver.Chrome(options = chrome_options)

  driver.get(urlGen(request.query))

  duration = 10

  sleeper = WebDriverWait(driver , duration)

  datas = driver.find_elements(By.XPATH ,'//*[@id="video-title"]')

  ids = []

  for data in datas:
      attr = data.get_attribute("href")
      if not shorts.match(str(attr)):
        id = videoId.findall(str(attr))
        if id:
          ids.append(id[0])

  for i in range(len(ids)):

    try:
      urllib.request.urlretrieve(imgUrl(ids[i]), f"{i}.jpg")
    except Exception as e:
      pass


  descriptions = []

  for i in range(len(ids)):
    descriptions.append(descriptionCleaner(pipe_Image2Text(imgToBase64(imgPath(i)))))


  sentiments = []

  for i in range(len(descriptions)):
    descriptions[i] = correctionCleaner(pipe_GrammarCorrection(descriptions[i]))
    sentiments.append(sentimentCleaner(pipe_SentimentAnalysis(descriptions[i])))


  pos_counter = neu_counter = neg_counter = 0

  pos_scores = neu_scores = neg_scores = []

  rating = 0

  output = {
      "label" : "",
      "rating" : 0
  }

  for sentiment in sentiments:

    match sentiment['label']:

      case 'positive':
        pos_counter +=1
        pos_scores.append(sentiment['score'])

      case 'neutral':
        neu_counter +=1
        neu_scores.append(sentiment['score'])

      case 'negative':
        neg_counter +=1
        neg_scores.append(sentiment['score'])


  max_occ = max(pos_counter , neu_counter , neg_counter)


  if max_occ == pos_counter:
    output['label'] = 'positive'
    output['rating'] =  np.round(mapper(np.median(pos_scores) , 1 , 5) , 1)
  elif max_occ == neu_counter:
    output['label'] = 'neutral'
    output['rating'] = np.round(mapper(np.median(neu_scores) , 1 , 5) , 1)
  else:
    output['label'] = 'negative'
    output['rating'] = np.round(mapper(np.median(neg_scores) , 1 , 5) , 1)

  return SentimentResponse(sentiment = output)
```

```python

ngrok_tunnel = ngrok.connect(8000)
print('Public URL:', ngrok_tunnel.public_url + '/docs')
nest_asyncio.apply()
uvicorn.run(app, port = 8000)
```

## StreamLit

```python
%%writefile main.py

import streamlit as st
import requests
from typing import List

def main():
    st.title("YT Thumbnail Analyzer")

    query = st.text_input("Enter query")

    if st.button("Analyze"):
        if not query:
            st.warning("Please enter a query")
        else:
            response = analyze_thumbnail_sentiment(query)
            st.header("Analysis")
            st.subheader(f"Sentiment : {response['sentiment']['label']}")
            st.subheader(f"Rating : {response['sentiment']['rating']}")

def analyze_thumbnail_sentiment(query: str) -> dict:
    api_url = "http://127.0.0.1:8000/analyze"
    payload = {"query": query}
    response = requests.post(api_url, json=payload)
    response.raise_for_status()
    return response.json()

if __name__ == "__main__":
  main()
```
